```python
import pandas as pd
import matplotlib.pyplot as plt

# Set a clean style for all our future graphs
plt.style.use('seaborn-v0_8-whitegrid')
print("Libraries imported successfully!")
```

```python
# 1. Load the data using Pandas
# Make sure the filename matches exactly what you uploaded
df_sales = pd.read_csv('Basic Problem 3 - Total sales per category.csv')

# 2. Select only the Top 10 categories to keep the chart clean
top_10_sales = df_sales.head(10)

# 3. Set up the visualization canvas
plt.figure(figsize=(12, 6))

# 4. Create a bar chart
plt.bar(top_10_sales['product_category'], top_10_sales['total_sales'], color='royalblue', edgecolor='black')

# 5. Add titles and labels to make the graph insightful
plt.title('Top 10 Product Categories by Total Sales', fontsize=16, fontweight='bold')
plt.xlabel('Product Category', fontsize=12, fontweight='bold')
plt.ylabel('Total Revenue', fontsize=12, fontweight='bold')

# 6. Formatting tweaks for readability
# Rotate the category names so they don't overlap
plt.xticks(rotation=45, ha='right')
# Prevent scientific notation (e.g., 1e6) on the y-axis
plt.ticklabel_format(style='plain', axis='y')

# Adjust layout so labels aren't cut off and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
df_orders = pd.read_csv('Intermediate Problem 1 - Number of orders per month in 2018.csv')

# 2. Map month numbers to names for a cleaner x-axis
month_names = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
               7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
df_orders['month_name'] = df_orders['order_month'].map(month_names)

# 3. Set up the canvas
plt.figure(figsize=(10, 5))

# 4. Create a line chart with markers for each data point
plt.plot(df_orders['month_name'], df_orders['total_orders'],
         marker='o', color='darkorange', linewidth=2, markersize=8)

# 5. Add titles and labels
plt.title('Monthly Order Trends in 2018', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Total Orders', fontsize=12, fontweight='bold')

# 6. Add grid lines to make the data points easy to read
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the correlation data
df_corr = pd.read_csv('Intermediate Problem 4 - Price Correlation.csv')

# 2. Set up the canvas
plt.figure(figsize=(10, 6))

# 3. Create the Scatter Plot
plt.scatter(df_corr['average_price'], df_corr['purchase_count'],
            alpha=0.5, color='seagreen', edgecolors='w')

# 4. Add titles and labels
plt.title('Correlation: Product Price vs. Number of Purchases', fontsize=16, fontweight='bold')
plt.xlabel('Average Price (per Product)', fontsize=12, fontweight='bold')
plt.ylabel('Total Purchases', fontsize=12, fontweight='bold')

# 5. Optional: Limit x-axis if you have extreme outliers (e.g., very expensive luxury items)
# This makes the main cluster of data easier to see
plt.xlim(0, df_corr['average_price'].quantile(0.95))

plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()

# 6. Calculate the actual correlation coefficient
correlation = df_corr['average_price'].corr(df_corr['purchase_count'])
print(f"The Pearson Correlation Coefficient is: {correlation:.2f}")
```
